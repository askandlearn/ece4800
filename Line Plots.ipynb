## ECE 4800: Line Plots in Python

Here we will look at how to plot lines using Python
* A 2D line is a function y = f(x)
* A 3D line is described by parametrized functions: x = f(t), y = g(t), and z = h(t)

Examples on how to plot these can be found at
* https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html
* https://matplotlib.org/mpl_toolkits/mplot3d/tutorial.html
* https://matplotlib.org/gallery/index.html

```python
# Import required libraries

# Numpy is python's library for vector math and MATLAB-like computation
import numpy as np 

# Matplotlib is python's plotting library
import matplotlib.pyplot as plt   
from matplotlib import cm

# Matplotlib's 3D plotting library
from mpl_toolkits.mplot3d import Axes3D 

# The notebook directive allows us to rotate 3D graphs, and save graphs
%matplotlib notebook  
```

```python
# Line in a 2 dimensional space
x = np.linspace(-1,1,40)
y = x**2
plt.figure()
plt.plot(x,y)  # If we just want a line, take out the o
plt.xlabel('X')
plt.ylabel('Y')
```

```python
# Line in a 3 dimensional space

# In this example, we draw a helix

# The helix is defined in terms of the parameter theta
# x = r*cos(theta)
# y = r*sin(theta)
# z = c*t

# Create a 3D axis
fig = plt.figure()
#ax = fig.gca(projection='3d')
ax = fig.add_subplot(111, projection='3d')

# Define the function in terms of the parameter theta over the range -4pi to 4pi
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
r = 2
x = r * np.sin(theta)
y = r * np.cos(theta)
z = 10*theta
ax.plot(x, y, z, label='helix')
ax.legend()
plt.show()
```
